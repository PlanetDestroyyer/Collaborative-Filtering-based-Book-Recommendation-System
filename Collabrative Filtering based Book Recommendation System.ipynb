```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
import requests
```

```python
# Define the URL of the CSV file
url = "https://raw.githubusercontent.com/MainakRepositor/Book-Recommender/master/data/BX-Books.csv"
# Fetch the content of the CSV file in chunks
response = requests.get(url, stream=True)

# Read the content in chunks and decode it using ISO-8859-1 encoding
chunks = []
for chunk in response.iter_content(chunk_size=8192):
    if chunk:
        chunks.append(chunk)
content = b''.join(chunks).decode('ISO-8859-1')

# Read the CSV content into a DataFrame
book = pd.read_csv(io.StringIO(content), sep='\t')
rating = pd.read_csv('Ratings.csv')
users = pd.read_csv('Users.csv')
```

```python
books_with_rating = rating.merge(book,on='ISBN')
```

```python
books_with_rating.head()
```

```python
num_rating = books_with_rating.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={"Book-Rating":"num_rating"},inplace=True)
```

```python
num_rating
```

```python
avg_rating = books_with_rating.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={"Book-Rating":"avg_rating"},inplace=True)
```

```python
avg_rating
```

```python
popular = num_rating.merge(avg_rating,on="Book-Title")
```

```python
popular
```

```python
popular[popular['num_rating']>=200]
```

```python
popular_df = popular[popular['num_rating']>=200].sort_values('avg_rating',ascending=False).head(100)
```

```python
popular_df = popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Publisher','num_rating','avg_rating','Image-URL-L']]
```

```python
popular_df.head()
```

```python
popular_df['avg_rating'] = popular_df['avg_rating'].apply(lambda x:round(x))
```

```python
popular_df
```

```python
x = books_with_rating.groupby('User-ID').count()['Book-Rating']>200
good_users = x[x].index
```

```python
filtered_rating = books_with_rating[books_with_rating['User-ID'].isin(good_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_book = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_rating
```

```python
final_rating.drop_duplicates()
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns="User-ID",values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0].shape
```

```python
def recommend(name):
    index = np.where(pt.index==name)[0][0]
    simlar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:10]
    data = []
    for i in simlar_items:
        item = []
        temp = book.loc[book["Book-Title"] == pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates('Book-Title', keep='first')['Book-Title']))
        item.extend(list(temp.drop_duplicates('Book-Title', keep='first')['Book-Author']))
        item.extend(list(temp.drop_duplicates('Book-Title', keep='first')['Image-URL-L']))

        
        
        
        data.append(item)
    return data
```

```python
recommend('The Notebook')
```
